```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from bcd import BayesianChangepointDetection
from bcd.distributions import StudentT
from bcd.distributions import Gaussian

import os
os.chdir('..')
```

```python
def hazard(p):
    def geometric(t):
        return p * (1 - p) ** t
    return geometric
```

```python
# Exponential distribution
data = np.concatenate((
    np.random.exponential(scale=1, size=100),
    np.random.exponential(scale=10, size=100)
    ))

# Student T
detector = BayesianChangepointDetection(hazard(0.6), StudentT(0.1, 0.1, 1, 1))
results = [detector.step(datum) for datum in data]

plt.plot([result.argmax() for result in results])
plt.plot(data)
plt.axvline(x=100, color='r')
plt.title('Student T')
plt.show()

# Gaussian
detector = BayesianChangepointDetection(hazard(0.6), Gaussian(0.1, 0.1, 1, 1))
results = [detector.step(datum) for datum in data]

plt.plot([result.argmax() for result in results])
plt.plot(data)
plt.axvline(x=100, color='r')
plt.title('Gaussian')
plt.show()
```
