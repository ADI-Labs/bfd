```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import math
```

```python
def test1():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    for t in np.arange(1, 484):
        X[t] = .75 * X[t-1] + e[t]
    for t in np.arange(484, 803):
        X[t] = 1.7 * X[t-1] - .75 * X[t-2] + e[t]
    for t in np.arange(803, 1024):
        X[t] = 1.25 * X[t-1] - .7 * X[t-2] + e[t]
    return X

plt.plot(test1())
plt.title("Test case 1")
plt.show()
```

```python
def test2():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    for t in np.arange(1, 576):
        X[t] = .75 * X[t-1] + e[t] + .75 * e[t-1]
    for t in np.arange(576, 879):
        X[t] = .75 * X[t-1] + e[t]
    for t in np.arange(879, 1024):
        X[t] = e[t] + .75 * e[t-1]
    return X

plt.plot(test2())
plt.title("Test case 2")
plt.show()
```

```python
def test3():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    for t in np.arange(1, 131):
        X[t] = .9 * X[t-1] + e[t]
    for t in np.arange(131, 1024):
        X[t] = .25 * X[t-1] + e[t]
    return X

plt.plot(test3())
plt.title("Test case 3")
plt.show()
```

```python
def test4():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    for t in np.arange(1, 421):
        X[t] = .75 * X[t-1] + e[t]
    # Diverges very quickly from this point onward
    for t in np.arange(421, 823):
        X[t] = 4 * (.75 * X[t-1] + e[t])
    for t in np.arange(823, 1024):
        X[t] = .75 * X[t-1] + e[t]
    return X

plt.plot(test4())
plt.title("Test case 4")
plt.show()
```

```python
def test5():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    a = 1 - .5 * np.cos(16 * math.pi * np.arange(1024) / 1024)
    for t in np.arange(1, 321):
        X[t] = a[t] * e[t]
    for t in np.arange(321, 665):
        X[t] = 5 * a[t] * e[t]
    # Diverges very quickly from this point onward
    for t in np.arange(665, 1024):
        X[t] = a[t] * X[t-1] + .81 * X[t-2] + e[t]
    return X

plt.plot(test5())
plt.title("Test case 5")
plt.show()
```

```python
def test6():
    X = np.zeros(1024)
    e = np.random.normal(size=1024)
    for t in np.arange(1, 470):
        X[t] = math.sqrt(abs(2 * e[t]))
    for t in np.arange(470, 1024):
        X[t] = .5 * X[t-1] + e[t]
    return X

plt.plot(test6())
plt.title("Test case 6")
plt.show()
```
